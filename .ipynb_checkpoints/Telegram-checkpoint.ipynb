```python
from cleantext import clean
from hazm import *
import pandas as pd
nrlmzr = Normalizer()
cleaned_df = pd.DataFrame(columns = ["message", "label"])

for i in range(len(df)):
  free_emoji_text = clean(df.iloc[i][0], no_emoji=True)
  normalized_text = nrlmzr.normalize(free_emoji_text)
  cleaned_df.loc[len(cleaned_df.index)] = [normalized_text, df.iloc[i][1]]
cleaned_df
```

```python
X = cleaned_df['message']
y = cleaned_df['label']
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.svm import SVC

svc = SVC(random_state=36)
svc.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = svc.predict(X_test)
print(classification_report(y_test, y_pred))
```
